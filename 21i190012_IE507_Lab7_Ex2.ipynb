##Ans 2 : (Observations)

We can see that the probability distribution graph constricts down as the no of bins is increased.

```python
import matplotlib.pyplot as plt

plot = sns.barplot(x="mpg",y="displacement",data=df)

h = plot.patches[0].get_height()
j = 0

for i in range(len(plot.patches)):
  if h < plot.patches[i].get_height():
    h = plot.patches[i].get_height()
    j = i

x=plot.patches[j].get_x()
y=h

plt.annotate('The highest bars in the barplot!', xy=(x,y), xytext=(x+0.3,y+55.5), arrowprops=dict(facecolor='red'))
```

##Ans 3 :

We can observe that as mpg increases displacement decreases.


```python
import numpy as np

md = sns.barplot(x="mpg",y="displacement",hue="model_year",data=df,estimator = np.median)
md.legend(bbox_to_anchor=(1.5, 1.0),loc='upper right')

h = md.patches[0].get_height()
j = 0

for i in range(len(md.patches)):
  if h < md.patches[i].get_height():
    h = md.patches[i].get_height()
    j = i

x=md.patches[j].get_x()
y=md.lines[j].get_ydata()[1]

plt.annotate(format((md.lines[j].get_ydata()[0]+md.lines[j].get_ydata()[1])/2,'.1f'), 
                   (x, y),
                   xytext = (x, y+20),
                   ha = 'center', va = 'top')
```

##Ans 4 :

We can observe that mpg is higher for later models and displacment is lower for them, whereas mpg is lower for old models and displacement higher.

```python
md = sns.barplot(x="mpg",y="displacement",hue="origin",data=df,estimator = np.median)
md.legend(bbox_to_anchor=(1.5, 1.0),loc='upper right')

h = md.patches[0].get_height()
j = 0

for i in range(len(md.patches)):
  if h > md.patches[i].get_height():
    h = md.patches[i].get_height()
    j = i

x=md.patches[j].get_x()
y=h

plt.annotate(format(h,'.1f'), 
                   (x, y),
                   xytext = (x, y+20),
                   ha = 'center', va = 'top')
```

##Ans 8 :

We can observe from the line representing the mean that y decreases with slight increases in between as x decreases.

```python
mywe = sns.boxplot(x="model_year",y="weight",data=df)
w_70 = []
w_74 = []

for i in range(len(df)):
  if df['model_year'][i]==70:
    w_70.append(df['weight'][i])
  elif df['model_year'][i]==74:
    w_74.append(df['weight'][i])

wp_70 = np.percentile(w_70, [75])
wp_74 = np.percentile(w_74, [75])

plt.annotate(format(wp_70[0],'.1f'), 
                   (0,wp_70[0]),
                   xytext = (0,wp_70[0]+200),
                   ha = 'center', va = 'top')

plt.annotate(format(wp_74[0],'.1f'), 
                   (4,wp_74[0]),
                   xytext = (4,wp_74[0]+200),
                   ha = 'center', va = 'top')
```
